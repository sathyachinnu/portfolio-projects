```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df =  pd.read_csv('Diwali Sales Data.csv' , encoding = 'unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.head()
```

## Sales based on Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender =df.groupby(['Gender'] ,as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sales_gender
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

## Sales based on age

```python
ax = sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age =df.groupby(['Age Group'] ,as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

## Orders based on States

```python
sales_state =df.groupby(['State'] ,as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state_o =df.groupby(['State'] ,as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state_o)
```

## Sales based on Marital Status

```python
ax = sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital =df.groupby(['Marital_Status','Gender'] ,as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sales_marital
```

```python
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',hue= 'Gender',data=sales_marital)
```

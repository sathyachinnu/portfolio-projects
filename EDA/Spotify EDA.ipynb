## Data analysis

## Top 5 popular artists

```python
top_five_artists = df.groupby("artist").count().sort_values(by='song_title',ascending = False)['song_title'][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

## Top 5 loudest tracks

```python
Top_five_loudest_tracks = df[['loudness','song_title']].sort_values(by='loudness')[:5]
Top_five_loudest_tracks                                                              
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness',y='song_title',data=Top_five_loudest_tracks)
plt.title('Top_five_loudest_tracks')

plt.show()
```

```python
df.columns
```

## Top five Artists by Danceablity

```python
Top_five_Artists_by_danceability= df[['danceability','song_title','artist']].sort_values(by='danceability',ascending = False)[:5]
Top_five_Artists_by_danceability
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='danceability',y='artist',data=Top_five_Artists_by_danceability)
plt.title('Top_five_Artists_by_danceability')
plt.show()
```

## Top 10 instrumentalness tracks

```python
Top_10_instrumentalness = df[['instrumentalness',"song_title", "artist"]].sort_values(by='instrumentalness',ascending =False)[:10]
Top_10_instrumentalness
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='instrumentalness',autopct='%1.2f%%',labels=Top_10_instrumentalness.song_title,data=Top_10_instrumentalness)
plt.title('Top_10_instrumentalness ')
plt.show()
```

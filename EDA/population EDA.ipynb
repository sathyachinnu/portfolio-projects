## Top 10 countries with highest population 2022

```python
top_10_countries=df[['Country','2022 Population']].sort_values(by='2022 Population',ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Country',y='2022 Population',data=top_10_countries)
plt.show()
```

## Correlation between columns

```python
dtpe = df.select_dtypes(include='number')
dtpe.head()
```

```python
dtpe.corr()
```

```python
sns.heatmap(dtpe.corr(), annot = True)

plt.rcParams['figure.figsize'] = (20,7)

plt.show()
```

## Population based on continent

```python

df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.histplot(data=df2)
```

```python
df2.plot()
```

```python
df3=df2.transpose()
df3
```

```python
plt.figure(figsize=(8, 6), dpi=80)
df3.plot()
```
